# Continuous Examples

```python
import pybees as pb
```

# [Levy](https://www.sfu.ca/~ssurjano/levy.html)

$f(\pmb{x^*}) = 0, at \pmb{x^*} = (1,..,1)$

```python
sbc = pb.SimpleBeesContinuous(
    n_scout_bees = 20, 
    elite_site_params = (5,15), 
    best_site_params = (5, 10),
    bounds = (-10,10), 
    n_dim = 2,
    nbhd_radius = 2,
)

sbc.optimize(pb.levy)
sbc.plot(global_min=(1,1))
```

# [Drop wave](https://www.sfu.ca/~ssurjano/drop.html)

$f(\pmb{x^*}) = -1, \ at \ \pmb{x^*} = (0,0)$

```python
sbc = pb.SimpleBeesContinuous(
    n_scout_bees = 30, 
    elite_site_params = (10, 40), 
    best_site_params = (10, 30),
    bounds = (-5.12,5.12), 
    n_dim = 2,
    nbhd_radius = 1
)
    
sbc.optimize(pb.drop_wave)
sbc.plot((0,0))
```

# [Michalewicz](https://www.sfu.ca/~ssurjano/michal.html)

$at\ d=2: f(\pmb{x^*}) = -1.8013, \ at \ \pmb{x^*} = (\ 2.20,\ 1.57)$ <br>
$at\ d=5: f(\pmb{x^*}) = -4.687658$ <br>
$at\ d=10: f(\pmb{x^*}) = -9.66015$

```python
sbc = pb.SimpleBeesContinuous(
    n_scout_bees = 15, 
    elite_site_params = (5, 30), 
    best_site_params = (5, 20),
    bounds = (-10,10), 
    n_dim = 2,
    nbhd_radius = 1,
)
    
sbc.optimize(pb.michalewicz)
sbc.plot()
```
